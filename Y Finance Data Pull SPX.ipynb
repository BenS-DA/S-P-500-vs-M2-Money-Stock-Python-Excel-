```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Download S&P 500 market data 

symbolframe = yf.download(['^GSPC'], start = '1980-11-03', end = '2025-12-12')
```

```python
# Examine the dataframe

symbolframe.info()
```

```python
# Inspect last ten rows of dataframe

symbolframe.tail(10)
```

```python
# Plot the dataframe

plt.figure(figsize = (10,5))
plt.plot(symbolframe['Close'],label='Close Price', color = 'green')
plt.title("Price")
plt.xlabel("Date")
plt.ylabel("Price in USD")
plt.legend()
plt.grid(True)
plt.show()
```

```python
symbolframe.to_csv('SPX.csv', index = True)
```
